### DDL Data Definition Language

DDL Data Definition Language: Used to deefine and manage structures

DDL: CREATE database objects and constraints

__For DDL DATABASEs there's ONLY 3 SQL statements that we can use!__

__CREATE__ : Create database objects

__ALTER___ : To modify the structure and/or characteristics of existing database objects

__DROP___ : to delete existing data base objects

```
CREATE TABLE Employee(
empId   Integer   NOT NULL,
empName   Char(25)  NOT NULL,
CONSTRAINT  empPk   PRIMARY KEY(empId)
);

CREATE TABLE Skill(
skillId   Integer   NOT NULL,
skillName Char(25)  NOT NULL,
CONSTRAINT  skillPk   PRIMARY KEY(skillId)
);

CREATE TABLE EmployeeSkill(
empId   Integer   NOT NULL,
skillId   Integer   NOT NULL,
skillLevel  Integer   NOT NULL,
CONSTRAINT  empSkillPk  PRIMARY KEY(empId, skillId),
CONSTRAINT  empFk   FOREIGN KEY(empId)    REFERENCES Employee(empId),
CONSTRAINT  skillFk   FOREIGN KEY(skillId)  REFERENCES Skill(skillId)
);

DROP TABLE EmployeeSkill;
```

__In case we need to update the information in all tables or delete__

```
CREATE TABLE EmployeeSkill(
empId   Integer   NOT NULL,
skillId   Integer   NOT NULL,
skillLevel  Integer   NULL,
CONSTRAINT  empSkillPk  PRIMARY KEY(empId, skillId),
CONSTRAINT  empFk   FOREIGN KEY(empId)    REFERENCES  Employee(empId) ON DELETE CASCADE,
CONSTRAINT  skillFk   FOREIGN KEY(skillId)  REFERENCES  Skill(skillId)  ON UPDATE CASCADE
);

```

### CREATING COLUMNS - Classification by CASE

```

# Classification by CASE : the CASE operator is like the IF statement in python and we can create a new column at the end with with AS

SELECT  type,
        heart_rate,
CASE
  WHEN  heart_rate  > 220-30
  THEN  'above max'
  WHEN  heart_rate  > ROUND(0.90 * (220-30))
  THEN  'above target'
  WHEN  heart_rate  > ROUND(0.50 * (220-30))
  THEN  'whithin target'
  ELSE  'below target'
END AS    'heartZone' # This is the NEW COLUMN
FROM    exercise_logs
;

SELECT    COUNT(*),
CASE
  WHEN  heart_rate  > 220-32
  THEN  'above max'
  WHEN  heart_rate  > ROUND(0.90 * (220-32))
  THEN  'above target'
  WHEN  heart_rate  > ROUND(0.50 * (220-32))
  THEN  'within target'
  ELSE  'below target'
END AS    'heartZone' # This is the NEW COLUMN
FROM    exercise_logs
GROUP BY  heartZone
;



SELECT    COUNT(*), # Attribute00
CASE
  WHEN  number_grade >= 90 AND number_grade <= 100
  THEN  'A'
  WHEN  number_grade >= 80 AND number_grade <90
  THEN  'B'
  WHEN  number_grade >= 70 AND number_grade <80
  THEN  'C'
  ELSE  'F'
END AS    'letter_grade' # Attribute01
FROM    student_grades
GROUP BY  letter_grade
;
```

### Retrieving Information from Multiple TABLES

Because using the SQL SELECT statement we exract DATA from only ONE TABLE and the result of this query is going to be a RELATION and because of this, we can use the results from a query as INPUT to another QUERY this is what we mean by the term SUBQUERIES!

__SUBQUERIES__ : Extract data from multiple TABLES

__SUBQUERY 1 : NON-CORRELATED__ : The Inner query only needs to run once in order to the DATABASE engine to solve the problems

__SUBQUERY 2 : CORRELATED__ :  The Inner query needs to run repeateadly in order for the DATABASE engine to solve the problem : where The INNER QUERY needs a value from the OUTTER QUERY in order to execute the INNER QUERY!!!

### NONCORRELATED SUBQUERY

```
SELECT  empName   # OUTER QUERY
FROM  Employee
WHERE deptId IN # Where the department Id is equal to the following set of values(The result of the INNER QUERY)
(
SELECT  deptId    # INNER QUERY : is the input for the WHERE statement
FROM  Department  # This TABLE *IS NOT* from the OUTTER TABLE! IMPORTANT!!!**********
WHERE depName LIKE 'Account%'
);
```

### CORRELATED SUBQUERY
```
SELECT  empName   # OUTER QUERY
FROM  Employee e  # e is and ALIAS for the Employee TABLE
WHERE empSalary >
(
SELECT AVG  (empSalary) # INNER QUERY
FROM    Employee  # This TABLE *IS* from the OUTTER TABLE! IMPORTANT!!!**********
WHERE   deptId = e.deptId # e is and ALIAS for the Employee TABLE
);
```


### RETRIEVING INFORMATION FROM MULTIPLE TABLES

DATA BASE JOINS : All DATABASE joins can be divided in two categories OUTER JOIN or INNER JOIN

__OUTER JOIN 1 : LEFT OUTER JOIN__

__OUTER JOIN 2 : RIGHT OUTER JOIN__

__OUTER JOIN 3 : FULL OUTER JOIN__

__INNER JOIN__ : The records only can be included in the result set if there are match pair of values in the two TABLES involved in the JOIN

__WHERE Clause INNER JOIN Example__

```
SELECT  empName, deptName
FROM  Employee AS E, Department AS D # Two tables!
WHERE   E.deptId = D.deptId   # Where TABLE matches! MATCHING PAIR OF VALUES! : The result is only going to be include if a MATCH occurs
;
```

#### JOIN ON

```
SELECT  empName, deptName
FROM  Employee AS E INNER JOIN Department AS D # INNER JOIN is *AND logic gate*
ON    E.deptId = D.deptId
WHERE D.deptName NOT LIKE 'Account%'
;
```


__OUTER JOIN__ : The records can be included in the result set if there are match pair of values in the two TABLES involved in the JOIN and also aditional records in the TABLE involved in the JOIN

__OUTER JOIN__ : syntax can be used to obtain data that exist in one table without matching data in the other table

__LEFT OUTER JOIN__


LEFT means give me a list of all LEFT TABLE records from Employee E. Therefore ALL LEFT DATA records must be include in the results. For this case the results will include ALL of the records from TABLE Employee E regardless whether they have matching records with the RIGHT TABLE Department D. but if they have a matching Department D include all matching deparment D as well in the result list

Give me a list of all of the employees regardles of whether they have a matching department but if they have a matching department include the deptName into the list

```
SELECT  empName, deptName
FROM  Employee E LEFT OUTER JOIN Department D
ON    E.deptId = D.deptId
;

```

__RIGHT OUTER JOIN__

RIGHT means give a list that icludes all RIGHT TABLE Department D records in the results regardless whether there are matching records with the left TABLE Employee E. but if there are matching employee inlude all the matching values from employeeName in the results.
 
```
SELECT  empName, deptName
FROM  Employee E RIGHT OUTER JOIN Department D
ON    E.deptId = D.deptId
;

```

__FULL OUTER JOIN__ : Add ALL RECORDS regardless matching values exists in both TABLES, but if a match exist all those matching records must be included.

```
SELECT  empName, deptName
FROM  Employee E FULL OUTER JOIN Department D
ON    E.deptId = D.depId
;
```
